```python
duckdb.sql("""
COPY (

    SELECT *
    FROM (

        SELECT
            *,
            DATE_TRUNC('month', trd_exctn_dt) AS mes_ref,

            ROW_NUMBER() OVER (
                PARTITION BY
                    cusip_id,
                    DATE_TRUNC('month', trd_exctn_dt)
                ORDER BY
                    trd_exctn_dt DESC
            ) AS rn

        FROM 'C:/Users/usuario/Desktop/data_ingestion/stage1_osbap_0k_volume_2025.parquet'

        WHERE trd_exctn_dt >= TIMESTAMP '2015-01-01'
          AND trd_exctn_dt < TIMESTAMP '2015-02-01'

    )

    WHERE rn = 1

)

TO 'C:/Users/usuario/Desktop/data_ingestion/teste_csv.csv'
(
    FORMAT CSV,
    HEADER TRUE
);
""")

print("Arquivo exportado com sucesso!")
```
