```python
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.functions import col, sum, count, avg, max, min, to_date
import matplotlib.pyplot as plt

# Criar sessão Spark
spark = SparkSession.builder \
    .appName("DataIngestion") \
    .config("spark.sql.parquet.inferTimestampNTZ.enabled", "false") \
    .config("spark.sql.legacy.parquet.nanosAsLong", "true") \
    .getOrCreate()
```

```python
df_spark = spark.read.parquet(
    r"C:\Users\usuario\Desktop\data_ingestion\main_panel_2025.parquet"
)
```

```python
df_spark = df_spark.withColumn(
    "date",
    (col("date") / 1_000_000_000).cast("timestamp")
)

df_spark = df_spark.withColumn(
    "date",
    to_date(col("date"))
)
```

```python
df_filtered = df_spark.filter(
    (col("date") >= "2015-01-01")
    & (col("country") == "USA")
)
```

```python
pdf = df_filtered.toPandas()

pdf.to_parquet(
    "df_filtered.parquet",
    engine="pyarrow",
    index=False
)
```

```python
import pandas as pd

df = pd.read_parquet(
    r"C:\Users\usuario\Desktop\data_ingestion\main_panel_2025.parquet"
)

df["date"] = pd.to_datetime(df["date"]).dt.date

df_filtered = df[
    (df["date"] >= pd.to_datetime("2010-01-01").date()) &
    (df["country"] == "USA")
]

df_filtered.to_parquet(
    r"C:\Users\usuario\Desktop\data_ingestion\df_filtered.parquet",
    engine="pyarrow",
    index=False
)

print("Arquivo exportado com sucesso!")
```
